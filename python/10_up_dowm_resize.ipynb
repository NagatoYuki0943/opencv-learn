```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/squirrel.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image.shape # h, w, c
```

# pyrUp 

```python
image_up = cv2.pyrUp(
    src=image,
    dstsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]w, [0]h
)
image_up.shape
```

```python
plt.imshow(image_up)
```

# pyrDown 

```python
image_down = cv2.pyrDown(
    src=image,
    dstsize=(image.shape[1] // 2, image.shape[0] // 2), # [1]w, [0]h
)
image_down.shape
```

```python
plt.imshow(image_down)
```

# resize 

interpolation:
- INTER_NEAREST       
- INTER_LINEAR        
- INTER_CUBIC         
- INTER_AREA          (,,,INTER_NEAREST)
- INTER_LANCZOS4      Lanczos4(8x8)
- INTER_MAX           
- WARP_FILL_OUTLIERS  ,,,0
- WARP_INVERSE_MAP    ,

```python
nearest = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]w, [0]h
    interpolation=cv2.INTER_NEAREST,
)
plt.imshow(nearest)
```

```python
linear = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]w, [0]h
    interpolation=cv2.INTER_LINEAR,
)
plt.imshow(linear)
```

```python
cubic = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]w, [0]h
    interpolation=cv2.INTER_CUBIC,
)
plt.imshow(cubic)
```

```python
area = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]w, [0]h
    interpolation=cv2.INTER_AREA,
)
plt.imshow(area)
```

```python
lanczos4 = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]w, [0]h
    interpolation=cv2.INTER_LANCZOS4,
)
plt.imshow(lanczos4)
```
