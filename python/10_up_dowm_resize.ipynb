```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/squirrel.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
image.shape # h, w, c
```

# pyrUp 上采样

```python
image_up = cv2.pyrUp(
    src=image,
    dstsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]代表w, [0]代表h
)
image_up.shape
```

```python
plt.imshow(image_up)
```

# pyrDown 下采样

```python
image_down = cv2.pyrDown(
    src=image,
    dstsize=(image.shape[1] // 2, image.shape[0] // 2), # [1]代表w, [0]代表h
)
image_down.shape
```

```python
plt.imshow(image_down)
```

# resize 多种差值算法

interpolation:
- INTER_NEAREST       最邻近插值
- INTER_LINEAR        线性插值
- INTER_CUBIC         双立方插值
- INTER_AREA          区域插值(使用像素区域关系的重采样,时图像抽取的首选方法,但是当图形被放大,它类似于INTER_NEAREST方法)
- INTER_LANCZOS4      Lanczos4(超过8x8邻域的插值算法)
- INTER_MAX           用于插值的掩模板
- WARP_FILL_OUTLIERS  标志位,用于填充目标图像的像素值,如果其中的一些值对应于原图像中的异常值,那么这些值将被设置为0
- WARP_INVERSE_MAP    标志位,反变换

```python
nearest = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]代表w, [0]代表h
    interpolation=cv2.INTER_NEAREST,
)
plt.imshow(nearest)
```

```python
linear = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]代表w, [0]代表h
    interpolation=cv2.INTER_LINEAR,
)
plt.imshow(linear)
```

```python
cubic = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]代表w, [0]代表h
    interpolation=cv2.INTER_CUBIC,
)
plt.imshow(cubic)
```

```python
area = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]代表w, [0]代表h
    interpolation=cv2.INTER_AREA,
)
plt.imshow(area)
```

```python
lanczos4 = cv2.resize(
    src=image,
    dsize=(image.shape[1] * 2, image.shape[0] * 2), # [1]代表w, [0]代表h
    interpolation=cv2.INTER_LANCZOS4,
)
plt.imshow(lanczos4)
```
