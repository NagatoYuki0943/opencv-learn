# line

```python
mat1 = cv2.line(
    mat.copy(),             # 要复制一份,否则修改原值
    pt1=(100, 150),         # point1 (x, y)
    pt2=(400, 450),         # point2 (x, y)
    color=(255, 0, 0),      # color
    thickness=3,            # 线宽
    lineType=cv2.LINE_AA    # LINE_4 LINE_8 LINE_AA
)
plt.imshow(mat1)
```

# getTextSize  获取文本尺寸参数

```python
text = "AbCdEfG"
fontsize = 3
thickness = 1
```

```python
((txt_w, txt_h), txt_bottom) = cv2.getTextSize(
    text=text,
    fontFace=cv2.FONT_HERSHEY_PLAIN,
    fontScale=fontsize,             # 字体相较于最初尺寸的缩放系数
    thickness=thickness,            # 字体笔画的粗细程度，默认值1
)
txt_w, txt_h, txt_bottom
```

```python
mat5 = cv2.rectangle(
    mat.copy(),                     # 要复制一份,否则修改原值
    pt1=(100, 400),                 # point1 (x, y)
    pt2=(100+txt_w, 400-txt_h-5),   # point2 (x, y)
    color=(50, 100, 100),           # color
    thickness=-1,                   # 线宽
    lineType=cv2.LINE_AA            # LINE_4 LINE_8 LINE_AA
)
plt.imshow(mat5)
```

# putText 添加文字
fontFace: 文字的字体类型（Hershey字体集），可供选择的有 <br/>
    FONT_HERSHEY_SIMPLEX：正常大小无衬线字体 <br/>
    FONT_HERSHEY_PLAIN：小号无衬线字体 <br/>
    FONT_HERSHEY_DUPLEX：正常大小无衬线字体，比FONT_HERSHEY_SIMPLEX更复杂 <br/>
    FONT_HERSHEY_COMPLEX：正常大小有衬线字体 <br/>
    FONT_HERSHEY_TRIPLEX：正常大小有衬线字体，比FONT_HERSHEY_COMPLEX更复杂 <br/>
    FONT_HERSHEY_COMPLEX_SMALL：FONT_HERSHEY_COMPLEX的小译本 <br/>
    FONT_HERSHEY_SCRIPT_SIMPLEX：手写风格字体 <br/>
    FONT_HERSHEY_SCRIPT_COMPLEX：手写风格字体，比FONT_HERSHEY_SCRIPT_SIMPLEX更复杂 <br/>
    这些参数和FONT_ITALIC同时使用就会得到相应的斜体字 <br/>

```python
mat5 = cv2.putText(
    mat5,                           # 要复制一份,否则修改原值
    text=text,                      # text
    org=(100, 400),                 # 要添加的文字基准点或原点坐标，左上角还是左下角取决于最后一个参数bottomLeftOrigin的取值
    fontFace=cv2.FONT_HERSHEY_PLAIN,# 文字的字体类型（Hershey字体集）
    fontScale=fontsize,             # 字体相较于最初尺寸的缩放系数
    color=(255, 0, 0),              # color
    thickness=thickness,            # 线宽
    lineType=cv2.LINE_AA,           # 字体笔画线条类型 LINE_4 LINE_8 LINE_AA
    bottomLeftOrigin=False,
)
plt.imshow(mat5)
```
