```python
import cv2
import imutils
import matplotlib.pyplot as plt
```

# OpenCV版本的检测
OpenCV 4发布之后，随着主要版本的更新，向后兼容性问题尤为突出。在使用OPenCV时，应检查当前正在使用哪个版本的OpenCV，然后使用适当的函数或方法。在imutils中的`is_cv2()、is_cv3()和is_cv4()`是可用于自动确定当前环境的OpenCV的版本简单的功能。

```python
print(cv2.__version__)
print(imutils.is_cv2())
print(imutils.is_cv3())
print(imutils.is_cv4())
```

```python
image = cv2.imread("../images/20230529-1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# 图像平移 translate

OpenCV中也提供了图像平移的实现，要先计算平移矩阵，然后利用仿射变换实现平移，在imutils中可直接进行图像的平移。

```python
image_t = imutils.translate(
    image=image,
    x=100,          # 沿x轴方向移动的像素个数
    y=50            # 沿y轴方向移动的像素个数
)
plt.imshow(image_t)
```

# 图像缩放 resize

图片的缩放在OPenCV中要注意确保保持宽高比。而在imutils中自动保持原有图片的宽高比，只指定宽度weight或Height即可

```python
image_re = imutils.resize(
    image=image,
    width=400,
    height=400,
)
plt.imshow(image_re)
```

# 逆时针图像旋转 rotate 裁切图片

在OpenCV中进行旋转时使用的是仿射变换，在这里图像旋转方法是`imutils.rotate()`

```python
image_ro = imutils.rotate(
    image=image,
    angle=30
)
plt.imshow(image_ro)
```

# 顺时针图像旋转 rotate_bound 缩小图片大小

```python
image_ror = imutils.rotate_bound(
    image=image,
    angle=30
)
plt.imshow(image_ror)
```

# 骨架提取 skeletonize
骨架提取，是指对图片中的物体进行拓扑骨架(topological  skeleton)构建的过程。

```python
skeleton = imutils.skeletonize(
    image=cv2.cvtColor(image, cv2.COLOR_RGB2GRAY),  # 必须是灰度图
    size=(3, 3) # 结构参数的尺寸(structuring  element)，相当于是一个粒度，越小需要处理的时间越长
)
plt.imshow(skeleton)
```
