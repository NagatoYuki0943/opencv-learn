```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

# opencv

## imread 默认读取全是3通道 RGB格式 要通过设置flags参数调整通道

```python
image1 = cv2.imread("../images/20230529.jpg")
print(image1.dtype)
print(image1.shape)
```

```python
# 读取灰度图也是三通道
image2 = cv2.imread("../images/gray.jpg")
print(image2.dtype)
print(image2.shape)
print(image2[0, 0]) # 三个通道值相同
```

```python
image3 = cv2.imread("../images/gray.jpg", flags=0)
print(image3.dtype)
print(image3.shape)
print(image3[0, 0]) # 三个通道值相同
```

## imwrite 保存图片

```python
cv2.imwrite("../images/20230529-1.jpg", image1)
```

## cvtColor 转换颜色

```python
plt.imshow(image1)
```

```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY))
```
