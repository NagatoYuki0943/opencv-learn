```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

## 1.创建空白图片

```python
image = np.zeros((400, 400)).astype(np.float32)
plt.imshow(image)
```

## 2.填充区域

```python
cv2.rectangle(
    img=image,
    pt1=(100, 100),
    pt2=(300, 300),
    thickness=-1,   # -1填充
    color=1.0
)
plt.imshow(image)
```

```python
c = cv2.Canny(
    image=(image * 255).astype(np.uint8),
    threshold1=0.5,
    threshold2=1,
    L2gradient=False,
)
plt.imshow(c)
```

## 3.找轮廓

```python
# 返回值为: 发现的轮廓对象, 图像的拓扑结构
contours, hierarchy = cv2.findContours(
    image=c,
    mode=cv2.RETR_TREE, # 方法
    method=cv2.CHAIN_APPROX_SIMPLE,
)
```

```python
# 测试一个点是否在给定的区域内部，边缘或者外部
p = []
for contour in contours:
    p.append(cv2.pointPolygonTest(contour=contour, pt=(10, 10), measureDist=True))
print(p)
```
