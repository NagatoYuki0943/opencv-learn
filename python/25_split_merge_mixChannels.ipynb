```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/94147214_p0.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# cv2.split 把多通道图像分为多个单通道图像

```python
r, g, b = cv2.split(m=image)
r.shape, g.shape, b.shape
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 8))
axes[0].imshow(r)
axes[0].axis(False)
axes[0].set_title("r")
axes[1].imshow(g)
axes[1].axis(False)
axes[1].set_title("g")
axes[2].imshow(b)
axes[2].axis(False)
axes[2].set_title("b")
```

## np 切片效果相同

```python
image.shape
```

```python
r1, g1, b1 = image[:, :, 0], image[:, :, 1], image[:, :, 2]
r1.shape, g1.shape, b1.shape
```

```python
# 效果相同
np.all(r==r1), np.all(g==g1), np.all(b==b1)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 8))
axes[0].imshow(r1)
axes[0].axis(False)
axes[0].set_title("r")
axes[1].imshow(g1)
axes[1].axis(False)
axes[1].set_title("g")
axes[2].imshow(b1)
axes[2].axis(False)
axes[2].set_title("b")
```

# cv2.merge 把多个单通道图像合并为多通道图像

```python
image1 = cv2.merge([r, g, b])
plt.imshow(image1)
```

## np.concatenate 效果相同

```python
image2 = np.stack([r1, g1, b1], axis=-1)
plt.imshow(image2)
```

# cv2.mixChannels 交换通道

```python
black = np.zeros(image.shape, dtype=np.uint8)
```

```python
cv2.mixChannels(
    src=image,
    dst=black,
    fromTo=[0, 2, 1, 1, 2, 0]
)
```
