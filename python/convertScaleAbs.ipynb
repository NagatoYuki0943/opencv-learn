```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# load image

```python
image = cv2.imread("../images/circle1.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(image.shape)
plt.imshow(image, cmap="gray")
```

```python
np.unique(image)
```

## THRESH_OTSU

```python
retval, binary = cv2.threshold(
    src=image,
    thresh=150, # 阈值
    maxval=255, # dst图像中最大值 只对 THRESH_BINARY 和 THRESH_BINARY_INV 有效果
    type=cv2.THRESH_BINARY_INV,
)
print(retval)
plt.imshow(binary, cmap="gray")
```

```python
np.unique(binary)
```

# Sober

```python
Gx = cv2.Sobel(
    src=binary,
    ddepth=cv2.CV_64F,  # 输出图像深度
    dx=1,       # X方向，几阶导数
    dy=0,       # Y方向，几阶导数
    ksize=3,    # SOBEL算子kernel大小，必须是奇数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_CONSTANT,
)
print(Gx)
print(np.unique(Gx))
plt.imshow(Gx)
```

# convertScaleAbs

```python
# 取像素绝对值
Gx_abs = cv2.convertScaleAbs(Gx)
print(np.unique(Gx_abs))
plt.imshow(Gx_abs)
```

```python
Gy = cv2.Sobel(
    src=binary,
    ddepth=cv2.CV_64F,  # 输出图像深度
    dx=0,       # X方向，几阶导数
    dy=1,       # Y方向，几阶导数
    ksize=3,    # SOBEL算子kernel大小，必须是奇数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_CONSTANT,
)
print(Gy)
print(np.unique(Gy))
plt.imshow(Gy)
```

```python
# 取像素绝对值
Gy_abs = cv2.convertScaleAbs(Gy)
print(np.unique(Gy_abs))
plt.imshow(Gy_abs)
```
