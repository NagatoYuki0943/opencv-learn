# getTextSize  获取文本尺寸参数

```python
text = "AbCdEfG"
fontsize = 3
thickness = 1
```

```python
((txt_w, txt_h), txt_bottom) = cv2.getTextSize(
    text=text,
    fontFace=cv2.FONT_HERSHEY_PLAIN,
    fontScale=fontsize,             # 字体相较于最初尺寸的缩放系数
    thickness=thickness,            # 字体笔画的粗细程度，默认值1
)
txt_w, txt_h, txt_bottom
```

```python
mat5 = cv2.rectangle(
    img=mat.copy(),                 # 要复制一份,否则修改原值
    pt1=(100, 400),                 # point1 (x, y)
    pt2=(100+txt_w, 400-txt_h-5),   # point2 (x, y)
    color=(50, 100, 100),           # color
    thickness=-1,                   # 线宽
    lineType=cv2.LINE_AA            # LINE_4 LINE_8 LINE_AA
)
plt.imshow(mat5)
```

# putText 添加文字
fontFace: 文字的字体类型（Hershey字体集），可供选择的有 <br/>
    FONT_HERSHEY_SIMPLEX：正常大小无衬线字体 <br/>
    FONT_HERSHEY_PLAIN：小号无衬线字体 <br/>
    FONT_HERSHEY_DUPLEX：正常大小无衬线字体，比FONT_HERSHEY_SIMPLEX更复杂 <br/>
    FONT_HERSHEY_COMPLEX：正常大小有衬线字体 <br/>
    FONT_HERSHEY_TRIPLEX：正常大小有衬线字体，比FONT_HERSHEY_COMPLEX更复杂 <br/>
    FONT_HERSHEY_COMPLEX_SMALL：FONT_HERSHEY_COMPLEX的小译本 <br/>
    FONT_HERSHEY_SCRIPT_SIMPLEX：手写风格字体 <br/>
    FONT_HERSHEY_SCRIPT_COMPLEX：手写风格字体，比FONT_HERSHEY_SCRIPT_SIMPLEX更复杂 <br/>
    这些参数和FONT_ITALIC同时使用就会得到相应的斜体字 <br/>

```python
mat5 = cv2.putText(
    img=mat5,                       # 要复制一份,否则修改原值
    text=text,                      # text
    org=(100, 400),                 # 要添加的文字基准点或原点坐标，左上角还是左下角取决于最后一个参数bottomLeftOrigin的取值
    fontFace=cv2.FONT_HERSHEY_PLAIN,# 文字的字体类型（Hershey字体集）
    fontScale=fontsize,             # 字体相较于最初尺寸的缩放系数
    color=(255, 0, 0),              # color
    thickness=thickness,            # 线宽
    lineType=cv2.LINE_AA,           # 字体笔画线条类型 LINE_4 LINE_8 LINE_AA
    bottomLeftOrigin=False,
)
plt.imshow(mat5)
```

# boundingRect 最小外接正矩形

```python
black = np.zeros((200, 200, 3), dtype=np.uint8)
# x, y
points = np.array([[50, 50], [160, 80], [30, 120], [140, 170]])
for point in points:
    cv2.circle(
        img=black,
        center=point,           # center (x, y)
        radius=1,               # 半径
        color=(255, 0, 0),      # color
        thickness=2,            # 线宽
    )
plt.imshow(black)
```

```python
# 返回左上角坐标和宽高
x1, y1, w, h = cv2.boundingRect(points)
x1, y1, w, h
```

```python
cv2.rectangle(
    img=black,
    pt1=(x1, y1),           # point1 (x, y)
    pt2=(x1 + w, y1 + h),   # point2 (x, y)
    color=(0, 255, 0),      # color
    thickness=1,            # 线宽为负值就填充
)
plt.imshow(black)
```

# minAreaRect 最小外接矩形

```python
# 最小外接矩形的中心(x，y)，(宽度，高度)，旋转角度(旋转角度是水平轴（x轴）逆时针旋转，与碰到的矩形的第一条边的夹角)
(x, y), (w, h), angle = cv2.minAreaRect(points=points)
x, y, w, h, angle
```

```python
t1 = cv2.rectangle(
    img=black.copy(),
    pt1=(int(x-0.5*w), int(y-0.5*h)),           # point1 (x, y)
    pt2=(int(x+0.5*w), int(y+0.5*h)),   # point2 (x, y)
    color=(0, 0, 255),      # color
    thickness=1,            # 线宽为负值就填充
)
plt.imshow(t1)
```

```python
# 将最小外接矩形的坐标转换为点坐标
minboxpoints = cv2.boxPoints(((x, y), (w, h), angle))
minboxpoints
```

```python
for i in range(4):
    pt1 = minboxpoints[i]
    pt2 = minboxpoints[(i + 1) % 4]

    cv2.line(
        img=black,
        pt1=[int(pt1[0]), int(pt1[1])],
        pt2=[int(pt2[0]), int(pt2[1])],
        color=(0, 0, 255),      # color
        thickness=1,            # 线宽
    )
plt.imshow(black)
```
