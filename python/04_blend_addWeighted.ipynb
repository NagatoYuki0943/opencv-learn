```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("../images/59021351_p0.jpg")
image2 = cv2.imread("../images/59880532_p0.jpg")
```

```python
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```

# addWeighted

- 参数1：src1，第一个原数组
- 参数2：alpha，第一个数组元素权重
- 参数3：src2第二个原数组
- 参数4：beta，第二个数组元素权重
- 参数5：gamma，图1与图2作和后添加的数值。不要太大，不然图片一片白。总和等于255以上就是纯白色了

```python
blend1 = cv2.addWeighted(image1, 0.5, image2, 0.5, 0)
blend1.shape
```

```python
plt.imshow(cv2.cvtColor(blend1, cv2.COLOR_BGR2RGB))
```

```python
blend2 = cv2.addWeighted(image1, 0.7, image2, 0.3, 0)
plt.imshow(cv2.cvtColor(blend2, cv2.COLOR_BGR2RGB))
```
