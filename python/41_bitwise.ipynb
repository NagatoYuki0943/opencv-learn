```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("../images/59021351_p0.jpg")
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(image1)
```

```python
image2 = cv2.imread("../images/59880532_p0.jpg")
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
```

```python
image1.shape, image1.dtype
```

```python
mask = np.zeros(image1.shape[:-1], dtype=np.uint8)
mask[:500] = 255
plt.imshow(mask, cmap="gray")
```

# bitwise_and

```python
b_and = cv2.bitwise_and(image1, image2, mask=mask)
plt.imshow(b_and)
```

# bitwise_or

```python
b_or = cv2.bitwise_or(image1, image2, mask=mask)
plt.imshow(b_or)
```

# bitwise_xor

```python
b_xor = cv2.bitwise_xor(image1, image2, mask=mask)
plt.imshow(b_xor)
```

# bitwise_not

```python
b_not = cv2.bitwise_not(image1, image2, mask=mask)
plt.imshow(b_not)
```
