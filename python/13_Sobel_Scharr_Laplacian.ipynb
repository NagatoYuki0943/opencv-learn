```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/squirrel.jpg")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

# Sobel

Sobel算子是离散微分算子（discrete differentiation operator），用来计算图像灰度的近似梯度

Sobel算子功能集合高斯平滑和微分求导

又被称为一阶微分算子，求导算子，在水平和垂直两个方向上求导，得到图像X方法与Y方向梯度图像

```python
s = cv2.Sobel(
    src=image,
    ddepth=-1,  # 输出图像深度
    dx=1,       # X方向，几阶导数
    dy=0,       # Y方向，几阶导数
    ksize=3,    # SOBEL算子kernel大小，必须是奇数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Sobel(
    src=image,
    ddepth=-1,  # 输出图像深度
    dx=0,       # X方向，几阶导数
    dy=1,       # Y方向，几阶导数
    ksize=3,    # SOBEL算子kernel大小，必须是奇数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Sobel(
    src=image,
    ddepth=-1,  # 输出图像深度
    dx=1,       # X方向，几阶导数
    dy=1,       # Y方向，几阶导数
    ksize=3,    # SOBEL算子kernel大小，必须是奇数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

# Scharr 比Sobel更多细节

```python
s = cv2.Scharr(
    src=image,
    ddepth=-1,  # 输出图像深度
    dx=1,       # X方向，几阶导数
    dy=0,       # Y方向，几阶导数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Scharr(
    src=image,
    ddepth=-1,  # 输出图像深度
    dx=0,       # X方向，几阶导数
    dy=1,       # Y方向，几阶导数
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Scharr(
    src=image,
    ddepth=-1,  # 输出图像深度
    dx=1,       # X,Y不能同时用
    dy=1,
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

# Laplacian

二阶导数计算Laplacian

解释：在二阶导数的时候，最大变化处的值为零即边缘是零值。通过二阶

导数计算，依据此理论我们可以计算图像二阶导数，提取边缘。

```python
l = cv2.Laplacian(
    src=image,
    ddepth=-1,  # 输出图像深度
    ksize=3,
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(l, cv2.COLOR_BGR2RGB))
```
