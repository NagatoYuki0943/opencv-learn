```python
import numpy as np
import cv2
from PIL import Image
import skimage
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../images/squirrel.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# Sobel

Sobeldiscrete differentiation operator

Sobel

XY

SobelcannyCanny
sobelSobel


```python
s = cv2.Sobel(
    src=image,
    ddepth=-1,  # 
    dx=1,       # X
    dy=0,       # Y
    ksize=3,    # SOBELkernel
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Sobel(
    src=image,
    ddepth=-1,  # 
    dx=0,       # X
    dy=1,       # Y
    ksize=3,    # SOBELkernel
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Sobel(
    src=image,
    ddepth=-1,  # 
    dx=1,       # X
    dy=1,       # Y
    ksize=3,    # SOBELkernel
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

# Scharr Sobel

```python
s = cv2.Scharr(
    src=image,
    ddepth=-1,  # 
    dx=1,       # X
    dy=0,       # Y
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
s = cv2.Scharr(
    src=image,
    ddepth=-1,  # 
    dx=0,       # X
    dy=1,       # Y
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(cv2.cvtColor(s, cv2.COLOR_BGR2RGB))
```

```python
# X,Y
s = cv2.Scharr(
    src=image,
    ddepth=-1,  # 
    dx=1,       # X,Y
    dy=1,
    scale=1,
    delta=0,
    borderType=cv2.BORDER_DEFAULT,
)
plt.imshow(s)
```
